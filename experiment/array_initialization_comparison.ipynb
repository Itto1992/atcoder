## 初期化
- 結論から言えば組み込みなら長さ1のリストを掛け算、numpyなら `np.zeros` , `np.ones` , `np.full` を使う

### 要素が全て$0$である$100\times100$の配列を初期化する

#### 組み込み

```python
%%time

table = []
for i in range(100):
    _table = []
    for j in range(100):
        _table.append(0)
    table.append(_table)
```

```python
%%time

table = [[0 for i in range(100)] for _ in range(100)]
```

```python
%%time

table = [[0]*100 for _ in range(100)]
```

```python
%%time

table = [[0]*100]*100
```

#### numpy

```python
%%time

import numpy as np

table  = np.zeros((100, 100))
```

```python
%%time

import numpy as np

table  = np.full((100, 100), 0)
```

### 要素が全て$-1$である$100\times100$の配列を初期化する

```python
%%time

table = []
for i in range(100):
    _table = []
    for j in range(100):
        _table.append(-1)
    table.append(_table)
```

```python
%%time

table = [[-1]*100 for _ in range(100)]
```

```python
%%time

table = [[-1 for i in range(100)] for _ in range(100)]
```

```python
%%time

import numpy as np

table  = np.ones((100, 100)) * -1
```

```python
%%time

import numpy as np

table  = np.full((100, 100), -1)
```

### 要素が全て$n$である$100\times100$の配列を初期化する

```python
n = 3
```

```python
%%time

table = []
for i in range(100):
    _table = []
    for j in range(100):
        _table.append(n)
    table.append(_table)
```

```python
%%time

table = [[n]*100 for _ in range(100)]
```

```python
%%time

table = [[n for i in range(100)] for _ in range(100)]
```

```python
%%time

import numpy as np

table  = np.ones((100, 100)) * n
```

```python
%%time

import numpy as np

table  = np.full((100, 100), n)
```

## アクセス速度
- 組み込みの方が早い
- numpyの場合は多重配列のindex指定を個別に行うよりまとめた方が早い

### 一つの要素にアクセス

```python
n = 0

table_py = [[n]*100 for _ in range(100)]
table_np  = np.full((100, 100), n)
```

```python
%%time 

table_py[50][50]
```

```python
%%time 

table_np[50][50]
```

```python
%%time 

table_np[50, 50]
```

### 複数の要素にスライスアクセス

```python
%%time 

table_py[:10][:10]
```

```python
%%time 

table_np[:10][:10]
```

```python
%%time

table_np[:10, :10]
```

## 結論
- 初期化速度はnumpy
- アクセス速度は組み込み

→アクセス回数が多いなら組み込み、少ないならnumpyがよい？

上記の$-1$で初期化する例だと50μs程度差があるので、単要素へのアクセスなら回数が50回以下ならnumpyでもペイできる？
